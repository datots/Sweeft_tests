```python
import collections
```

```python
D = int(input())
m = collections.OrderedDict()

for j in range(D):
    word = input()
    if word in m:
        m[word] += 1
    elif len(m) > 160:
        break
    else:
        m[word] = 1

print(len(m))

for c,x in m.items():
    print(x, end = " ")
```
