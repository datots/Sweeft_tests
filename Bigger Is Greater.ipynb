```python
def bigger_Is_Greater(s):
    arr = list(s)
   
    l = len(arr) - 1
    while l > 0 and arr[l - 1] >= arr[l]:
        l -= 1
    if l <= 0:
        return "no answer"
    
    
    k = len(arr) - 1
    while arr[k] <= arr[l - 1]:
        k -= 1
    arr[l - 1], arr[k] = arr[k], arr[l - 1]
    
    
    arr[l : ] = arr[len(arr) - 1 : l - 1 : -1]
    return "".join(arr)

for _ in range(int(input())):
    print(bigger_Is_Greater(input()))
```
