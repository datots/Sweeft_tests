```python
def boom(net):
    outcome = [list('O' * len(row)) for row in net]
    for y, row in enumerate(net):
        for g, x in enumerate(row):
            if x == 'O':
                outcome[y][g] = '.'
                if 0 < y:
                    outcome[y-1][g] = '.'
                if y+1 < len(net):
                    outcome[y+1][g] = '.'
                if 0 < g:
                    outcome[y][g-1] = '.'
                if g+1 < len(net[y]):
                    outcome[y][g+1] = '.'
    return [''.join(row) for row in outcome]
     

def bomber_Man(n, net):
    
    if n == 1:
        return net
    
    if n % 2 == 0:
        return ['O' * len(row) for row in net]
    elif (n+1) % 4 == 0:
        return boom(net)
    elif (n-1) % 4 == 0:
        return boom(boom(net))
    else:
        raise ValueError

if __name__ == "__main__":
    r, c, n = input().strip().split(' ')
    r, c, n = [int(r), int(c), int(n)]
    net = []
    net_i = 0
    for net_i in range(r):
        net_t = str(input().strip())
        net.append(net_t)
    outcome = bomber_Man(n, net)
    print ("\n".join(map(str, outcome)))


```
